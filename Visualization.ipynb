```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import os
import zipfile
from google.colab import files

df = pd.read_csv('relative_entropy.txt', encoding='utf-16', header=None, names=['play1', 'play2', 'relative_entropy'])

relative_entropy = df.pivot(index='play1', columns='play2', values='relative_entropy')
relative_entropy.fillna(0, inplace=True)

index_order = df['play1'].unique()
column_order = df['play2'].unique()

last_column = column_order[-1]
column_order = [last_column] + list(column_order[:-1])

relative_entropy = relative_entropy.reindex(index=index_order, columns=column_order)

relative_entropy = relative_entropy.rename_axis(None).rename_axis(None, axis=1)
relative_entropy = relative_entropy.round(2)

display(relative_entropy)
```

```python
print(relative_entropy.values.mean())
```

```python
plt.figure(figsize=(30, 25))
sns.heatmap(relative_entropy, annot=True, cmap='viridis', fmt='.2f', linewidths=.5)
plt.title("Heatmap of relative entropies between William Shakespeare's plays")
plt.show()
```

```python
relative_entropy = relative_entropy.replace(0, np.nan)

fig = plt.figure(figsize=(30, 30))
ax = fig.add_subplot(111, projection='3d')

x_labels = relative_entropy.columns.values
y_labels = relative_entropy.index.values

for i in range(len(y_labels)):
    row_data = relative_entropy.loc[y_labels[i]].values
    ax.plot(np.arange(len(x_labels)), np.full(len(x_labels), i), row_data, marker='o')

ax.set_xticks(np.arange(len(x_labels)))
ax.set_xticklabels(x_labels)
ax.set_yticks(np.arange(len(y_labels)))
ax.set_yticklabels(y_labels)
ax.set_xlabel('play1')
ax.set_ylabel('play2')
ax.set_zlabel('relative_entropy')
ax.set_title("3D Plot of relative entropies between William Shakespeare's plays")

ax.view_init(30, 30)

plt.show()
```

```python
zip_filename = '/plots.zip'

list = relative_entropy.index.tolist()

with zipfile.ZipFile(zip_filename, 'a', zipfile.ZIP_DEFLATED, False) as zip_file:
    for idx in relative_entropy.index:
        y_values = relative_entropy.loc[idx]
        x_values = relative_entropy.columns
        mask = x_values != idx

        if mask.sum() > 0:
            plt.figure(figsize=(10, 6))
            plt.plot(x_values[mask], y_values[mask], marker='o')
            plt.xlabel('play_name')
            plt.ylabel('relative_entropy')
            plt.title(f'play name: relative entropy of {idx}')
            plt.legend(title=idx, bbox_to_anchor=(1.05, 1), loc='upper left')
            plt.xticks(rotation=90)
            plt.grid(True)
            plt.tight_layout()

            plot_filename = f'/similarity_plot_{idx}.png'
            plt.savefig(plot_filename)
            plt.close()

            zip_file.write(plot_filename)

            os.remove(plot_filename)

        list.remove(idx)

from google.colab import files
files.download(zip_filename)
```
